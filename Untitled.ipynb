```python
def line_x(x1, x2, x_fin):
    y_diff = x2[1] - x1[1]
    x_diff = x2[0] - x1[0]
    try:
        m = y_diff/x_diff
    except:
        #print("Straight line")
        return x_fin[0]
    x = ((-x_fin[1])/m) + x_fin[0]
    return x    
```

```python
def line(x1, x2, y):
    y_diff = x2[1] - x1[1]
    x_diff = x2[0] - x1[0]
    try:
        m = y_diff/x_diff
    except:
        #print("Straight line")
        return x1[0]
    x = ((y-x1[1])/m) + x1[0]
    return x    
```
